```python
# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS
```

download file alice_novel https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/alice_novel.txt 
download file image alice_novel https://github.com/ardhiraka/PFDS_sources/raw/master/alice_mask.png


```python
# open the file and read it into a variable alice_novel
alice_novel = open('data/alice_novel.txt', 'r').read()
```

```python
alice_novel
```

```python
stopwords = set(STOPWORDS)
stopwords
```

Buatword cloud object dan generate word cloud. Agar lebih ringkas, mari kita buat word cloud hanya dengan menggunakan 2000 kata pertama dalam novel.

```python
alice_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)
```

```python
# generate the word cloud
alice_wc.generate(alice_novel)
```

```python

import matplotlib.pyplot as plt 
```

```python
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Kita perbesar gambar tersebut agar lebih jelas menggunkan figure 

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# display the cloud
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# stopwords.update(["drink", "now", "wine", "flavor", "flavors"])

#stopwords.add('said') # add the words said to stopwords
stopwords.update(["said", "one", "go"])

# re-generate the word cloud
alice_wc.generate(alice_novel)

# display the cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import numpy as np
from PIL import Image # converting images into arrays
```

```python
alice_mask = np.array(Image.open('data/alice_mask.png'))
```

```python
alice_mask 
```

```python
alice_mask.shape # lihat gambar tersebut 
```

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Membentuk word cloud sesuai dengan masking menggunakan word_cloud . 
Kita tetap menggunakan 2000 kata pertama dalam novel.

```python
stopwords ## sdh diupdate sebelumnya 
```

```python

# instantiate a word cloud object
alice_wc = WordCloud(
            background_color='white', 
            max_words=2000, 
            mask=alice_mask, 
            stopwords=stopwords)
```

```python
# generate the word cloud
alice_wc.generate(alice_novel)

# display the word cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
