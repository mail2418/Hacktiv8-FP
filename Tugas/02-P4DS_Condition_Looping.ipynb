## Python "while" Loops

    Iteration artinya mengeksekusi blok kode yang sama berulang kali, berpotensi berkali-kali. Struktur pemrograman yang mengimplementasikan iterasi disebut loop.

Dalam pemrograman, ada dua jenis iterasi, indefinite dan definite:
* Dengan indefinite iteration, 
    berapa kali loop dijalankan tidak ditentukan secara eksplisit sebelumnya. Sebaliknya, blok yang ditunjuk dijalankan berulang kali selama beberapa kondisi terpenuhi.
* Dengan definite iteration, 
    berapa kali blok yang ditunjuk akan dieksekusi ditentukan secara eksplisit pada saat perulangan dimulai.
    
Format while loop ditampilkan di bawah ini:


````python
while <expr>:
    <statement(s)>
````

````python
x = 10 
y = 11 
while x < y: 
    print("Helo looping terus")
````

````python
n = 5
while n > 0:
    print(n)
    n -= 1  # n = n - 1 
   
````

````python
i = 1
while i <= 6:
  print(i)
  i += 1
````

````python
ok = True
x = 0  
while ok: 
    print("Hello")
    if x == 2:  
        ok = False
    x += 1 
````

### The Python break and continue Statements

Perbedaan antara break dan continue ditunjukkan pada diagram berikut:



break and continue statement 

![alt text](https://i.ibb.co/QFLdx58/breakcontinue.png "Grouping statement")



````python
i=0
while(i<8):
    i+=1
    if(i==6): continue
    print(i)
````

````python
n = 5
while n > 0:                 # 5 > 0  : true        #  4 > 0 
    n -= 1  # 4 , 3          # 5 - 1 = 4            #  4 - 1 = 3 
    if n == 3:               # 4 == 3 : false       #. 3 == 3 : true 
        break # Break Statement   # escape          #    break 
    print(n)                 # print(n) = 4      
print('Loop ended.')

 
````

````python
n = 10
while n > 0:                   # 5 > 0  : true      # 4 > 0              # 3 > 0                  # 2 >0 
    n -= 1                     # 5 - 1 = 4          # 4 -1 = 3           # 3 -1 = 2               # 2-1 = 1
    if n % 2 == 0:             # if 4 % 2 == 0 : True  # if 3 % 2 == 0 : # if 2 % 2 == 0 : true   # 1 % 2 == 0  : false  
        continue               #   continue                              #   continue            
    print(n)                                        # print(3)                                      # print(1)
print('Loop ended.')
````

```
while <expr>:
    <statement(s)>
else:
    <additional_statement(s)>
```    

````python
n = 5
while n > 0:
    n -= 1
    print(n) 
    x = n +1 
else:
    print('Loop done.')
    y = 0 

print(x)    
print(y)
````

````python
age = 65 
gender = 'F'

if age < 18:
    if gender == 'M':
        print('son')
    else:
        print('daughter')
elif age >= 18 and age < 65:    # age >= 18  = True  and  False 
    if gender == 'M':
        print('father')
    else:
        print('mother')
else:
    if gender == 'M':
        print('grandfather')
    else:
        print('grandmother')
````
