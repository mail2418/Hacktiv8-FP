# Forecasting

```python
df_td_subset_s.index[-1]
```

```python
n_periods = 30*3
s_fitted, confint = smodel.predict(n_periods=n_periods,return_conf_int=True)
index_of_fc = pd.date_range(df_td_subset_s.index[-1], periods=n_periods, freq="D")

fitted_series = pd.Series(s_fitted.values, index=index_of_fc)
lower_series = pd.Series(confint[:, 0], index=index_of_fc)
upper_series = pd.Series(confint[:, 1], index=index_of_fc)
```

```python
fig2 = plt.figure(figsize=(20,10), constrained_layout = True)
axs = fig2.subplot_mosaic(
    """
    AB
    CC
    """
)
fig2.suptitle("TD Stock Price Analysis", weight="bold", size=24)
axs['A'].plot(df_td_date[6020:6520], df_td['close'][6020:6520], label="TD", c="blue", linewidth=1.5)
axs['A'].plot(df_ry_date[6240:6740], df_ry['close'][6240:6740], label="RY", c="yellow", linewidth=1.5, alpha=0.75)
axs['A'].plot(df_bns_date[5250:5750], df_bns['close'][5250:5750], label="BNS", c="black", linewidth=1.5, alpha= 0.8)

axs['A'].set_xlabel('Date', weight="bold", size=18)
axs['A'].set_ylabel("Price (CAD)", weight="bold", size=18)
axs['A'].set_title("Closing Price Comparison", weight="bold", size=20)
axs['A'].legend(fontsize="large", title="Canadian Banks", title_fontsize=12)
for label in (axs['A'].get_xticklabels() + axs['A'].get_yticklabels()):
    label.set_fontsize(16)

axs['B'].plot(df_td_date[6220:6520], new_df_td['close'][6220:6520], label='Close Price', color='#0D22D6', alpha=1, linewidth=1.5)
axs['B'].plot(df_td_date[6220:6520], new_df_td['Moving10'][6220:6520], label='10-day Average', linestyle='--', color='orange', alpha=1.0, linewidth=1.5)
axs['B'].plot(df_td_date[6220:6520], new_df_td['Moving30'][6220:6520], label='30-day Average', linestyle='--', color='#C50AFC', alpha=0.95, linewidth=1.5)

axs['B'].set_xlabel('Date', weight="bold", size=18)
axs['B'].set_ylabel('Price (CAD)', weight="bold", size=18)
axs['B'].set_title('TD Close Moving Average', weight="bold", size=20)
axs['B'].legend(fontsize="large", title="Period", title_fontsize=12, loc='upper left')
axs['B'].axvspan(*mdates.datestr2num(['8/5/2021','8/30/2021']), color='#0DD622', alpha=0.15)
axs['B'].axvspan(*mdates.datestr2num(['3/20/2022','4/15/2022']), color='#0DD622', alpha=0.15)
axs['B'].axvspan(*mdates.datestr2num(['5/25/2022','6/20/2022']), color='#0DD622', alpha=0.15)

for label in (axs['B'].get_xticklabels() + axs['B'].get_yticklabels()):
    label.set_fontsize(16)

axs['C'].plot(df_td_subset_s.index[125:], df_td_subset_s.close[125:], label='Observed', color='#0D22D6', alpha=1, linewidth=1.5)
axs['C'].plot(fitted_series, linestyle='--', label='Forecast', color='#F0083D', linewidth=1.5)
axs['C'].fill_between(lower_series.index, 
                 lower_series, 
                 upper_series, 
                 color='k', alpha=.15)
axs['C'].set_xlabel('Date', weight="bold", size=18)
axs['C'].set_ylabel('Price (CAD)', weight="bold", size=18)
axs['C'].legend(loc='upper left',fontsize='large')
axs['C'].set_title("SARIMA - TD Closing Price Forecasting in 3 months", weight="bold", size=20 )

for label in (axs['C'].get_xticklabels() + axs['C'].get_yticklabels()):
    label.set_fontsize(16)

plt.show()
```

```python
from sklearn.model_selection import TimeSeriesSplit
```

```python
tss = TimeSeriesSplit()
```

```python
tss.split(df_td[365*7:])
```

```python
fig,axs = plt.subplots(5,1, figsize=(20,10), constrained_layout=True)
fig.suptitle("Data Train/Test Split", weight="bold", size=14)
fold=0
for train,test in tss.split(df_td[365*7:]):
    train_data = df_td[365*7:].iloc[train]
    test_data = df_td[365*7:].iloc[test]
    train_data['close'].plot(ax=axs[fold], label="Training Set", color="blue", linewidth=1.5, alpha=0.8)
    test_data['close'].plot(ax=axs[fold], label="Test Set", color="orange", linewidth=1.5)
    axs[fold].axvline(test_data.index.min(), color="black", linestyle="--")
    axs[fold].set_title(f"Fold {fold}", size=12, weight="bold")
    for label in (axs[fold].get_xticklabels() + axs[fold].get_yticklabels()):
        label.set_fontsize(12)
    fold +=1

axs[4].legend(loc="upper center")
axs[4].set_xlabel("Date",size=12,weight="bold")
plt.show()


```
