# Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
new_dataset['NMHC(GT)'] = new_dataset['NMHC(GT)'].apply(lambda x: float(x))
```

```python
col = new_dataset.columns.tolist()
type(col)
```

```python
X = new_dataset[col].drop('T',1)
y = new_dataset['T']
```

```python
X = X.drop(columns=['Date','Time'])
```

```python
# dataset['hour'] = dataset["Time"].apply(lambda x: int(x.split(':')[0]))
# dataset.hour.head()
```

```python
X
```

```python
ss= StandardScaler()
X_ss = ss.fit_transform(X)
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_ss,y, test_size=0.1, random_state=42)
```

```python
model_lg = LinearRegression()
```

```python
model_lg.fit(X_train, y_train)
model_lg.intercept_
```

```python
predict = model_lg.predict(X_test)
predict
```

```python
from sklearn import metrics
```

```python
all_metrics = []
```

```python
rs = metrics.mean_squared_error(y_test,predict)
all_metrics.append(rs)
rs
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
```

```python
model = LogisticRegression()
lab_enc = LabelEncoder()
```

```python
training_y = lab_enc.fit_transform(y_train)
model.fit(X_train, training_y)
```

```python
predict = model.predict(X_test)
predict
```

```python
rs = metrics.mean_squared_error(y_test,predict)
all_metrics.append(rs)
rs
```

# Decsion Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model_DTR = DecisionTreeRegressor()
model_DTR.fit(X_train, y_train)
```

```python
predict = model_DTR.predict(X_test)
predict
```

```python
rs = metrics.mean_squared_error(y_test, predict)
all_metrics.append(rs)
rs
```

# SVM

```python
from sklearn.svm import SVR
```

```python
model_svr = SVR()
model_svr.fit(X_train, y_train)
```

```python
predict = model_svr.predict(X_test)
predict
```

```python
rs = metrics.mean_squared_error(y_test,predict)
all_metrics.append(rs)
rs
```

```python
all_models = ['LinearRegression', 'Logistic Regression', 'Decision Tree', 'SVR']
compare_model_df = pd.DataFrame({'Models': all_models, 'Metrics': all_metrics})
```
